```python
import requests

def get_tradable_coins():
    url = 'https://api.kraken.com/0/public/AssetPairs'
    response = requests.get(url)
    data = response.json()

    if data['error']:
        print("Error:", data['error'])
        return []

    tradable_coins = []
    for pair in data['result']:
        base = data['result'][pair]['base']
        quote = data['result'][pair]['quote']
        tradable_coins.append((base, quote))

    return tradable_coins

coins = get_tradable_coins()
```

# Historical

```python
import requests

def get_ohlc_data(pair, interval=60):
    url = f'https://api.kraken.com/0/public/OHLC?pair={pair}&interval={interval}'
    response = requests.get(url)
    data = response.json()

    if data['error']:
        print("Error:", data['error'])
        return []

    # Find the correct key for the asset pair
    result_key = list(data['result'].keys())[0]
    ohlc_data = data['result'][result_key]
    return ohlc_data

# Example usage
pair = 'XBTUSD'  # Bitcoin to USD
interval = 60  # 1-hour interval
ohlc_data = get_ohlc_data(pair, interval)
```

# More historical data

```python
import requests
import time

def get_ohlc_data(pair, interval=60, since=None):
    url = f'https://api.kraken.com/0/public/OHLC?pair={pair}&interval={interval}'
    if since:
        url += f'&since={since}'
    response = requests.get(url)
    data = response.json()

    if data['error']:
        print("Error:", data['error'])
        return []

    result_key = list(data['result'].keys())[0]
    ohlc_data = data['result'][result_key]
    return ohlc_data

# Example usage
pair = 'XBTUSD'  # Bitcoin to USD
interval = 60  # 1-hour interval
since = int(time.time()) - 3 * 24 * 60 * 60  # 3 days ago
ohlc_data = get_ohlc_data(pair, interval, since)
```
